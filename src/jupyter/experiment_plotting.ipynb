```python
stds_train = []
accs_train = []
stds_test = []
accs_test = []
step = 10
for i in range(0, len(df), step):
    # print(df[i - 10:i:][])
    stds_train.append(df[i:i+step:]["BoostAccuracy_accuracy"].std())
    accs_train.append(df[i:i + step:]["BoostAccuracy_accuracy"].mean())
    stds_test.append(df[i:i + step:]["BoostAccuracyTest_accuracy"].std())
    accs_test.append(df[i:i + step:]["BoostAccuracyTest_accuracy"].mean())
```

```python
fig = go.Figure()
df_t = df[:10:]
def plot_boosting_train_test_acc(ddf: pd.DataFrame):
    fig.add_trace(go.Scatter(x=np.arange(0, len(stds_train)), y=accs_train,
                        mode='lines',
                        name='BoostAccuracyTrain',
                        error_y=dict(
                                     type='data',  # value of error bar given in data coordinates
                                     array=stds_train,
                                     visible=True)
                                 ))
    
    fig.add_trace(go.Scatter(x=np.arange(0, len(stds_test)), y=accs_test,
                        mode='lines',
                        name='BoostAccuracyTest',
                        error_y=dict(
                                     type='data',  # value of error bar given in data coordinates
                                     array=stds_test,
                                     visible=True)
                                 ))
    
    
    fig.update_layout(height = 500)
    fig.update_xaxes(range=[-0.5, 10])
    fig.update_layout(legend_title_text="Boosting accuracy over 10 runs <br>  with 10-fold cross-validation")
    fig.show()
plot_boosting_train_test_acc(df)
```

```python

# with open("./../../ais/logg_dat/logg_dat_PrimaDiabetes.json", "r") as f:
# with open("./../../ais/logg_dat/logg_dat_Diabetes.json", "r") as f:
# with open("./../../ais/logg_dat/logg_dat_Sonar.json", "r") as f:
# with open("./../../ais/logg_dat/logg_dat_Glass.json", "r") as f:
# with open("./../../ais/logg_dat/logg_dat_Ionosphere.json", "r") as f:
# with open("./../../ais/logg_dat/logg_dat_Wine.json", "r") as f:
#with open("./../../ais/logg_dat/logg_dat_EmbeddingBuzfeed.json", "r") as f:
file_pre = "ex_boosting/logg_dat_PrimaDiabetes_boosting"
with open(f"./../../ais/logg_dat/{file_pre}.json", "r") as f:


#with open("./../../ais/logg_dat/logg_dat_Iris.json", "r") as f:
    js_obj = json.load(f)

df_list = parse_iter_props(js_obj)
sample_rate = 1
for df in df_list:
    gen_cor_w_noreg_cols(df, "TestAccuracy")
    gen_cor_w_noreg_cols(df, "TrainAccuracy")
    gen_cor_w_noreg_cols(df, "BoostAccuracy")
    gen_cor_w_noreg_cols(df, "BoostAccuracyTest")
    df.dropna(how="all", inplace=True)
    df.reset_index(drop=True, inplace=True)

def plot_train_test_acc(ddf: pd.DataFrame, n: int):
    markers = ["circle", "square", "diamond", "cross", "star"]
    fig = go.Figure()
    # ddf = ddf.copy().dropna(how="all").reset_index(drop=True)
    fig.add_trace(go.Scatter(x=ddf[df.index % sample_rate == 0].index, y=ddf.iloc[::sample_rate, :]["TestAccuracy_accuracy"].ffill(),
                        mode='lines+markers',
                        name='Test Accuracy',
                        marker=dict(
                            symbol=markers[0],
                            size=8,
                            #angleref="previous",
                            )))
    fig.add_trace(go.Scatter(x=ddf[df.index % sample_rate == 0].index, y=ddf.iloc[::sample_rate, :]["TrainAccuracy_accuracy"].ffill(),                        mode='lines+markers',
                        name='Train Accuracy',
                        marker=dict(
                            symbol=markers[1],
                                size=8,
                                #angleref="previous",
                            )))
    
    fig.update_layout(height = 500)
    prefix = "accuracy"
    path = Path(f"../plots/{file_pre}/{prefix}")
    if not path.is_dir():
        path.mkdir(parents=True)
    fig.write_image(f"../plots/{file_pre}/accuracy/fold_{n}.png", scale=2)
    fig.show()
    
def plot_score_comp_memberships(ddf: pd.DataFrame, fold: int, prefix: str):
    # ddf = ddf.copy().dropna(how="all").reset_index(drop=True)
    cols = ddf.columns
    fig = go.Figure()
    markers = ["circle", "square", "diamond", "cross", "star", "triangle-up", "triangle-down", "triangle-left", "triangle-right"]
    n = 0
    for col in cols:
        if col.startswith(prefix):
            fig.add_trace(go.Scatter(x=ddf.index, y=ddf[col].ffill(),
                                mode='lines+markers',
                                name=col[len(prefix):],
                                marker=dict(
                                            symbol=markers[n],
                                            size=8,
                                            #angleref="previous",
                                        )))
            n += 1


    write_image(fig, f"{file_pre}/{prefix}", f"fold_{fold}.png")
    fig.show()
    
# plot_agg_vals(df_list, "PopLabelMemberships_", file_pre, sample_rate, 50)
fig_agg_vals = plot_agg_vals(df_list, "ScoreComponents_", file_pre, sample_rate)
fig_agg_vals.update_yaxes(range=[-1, 1])
fig_agg_vals.update_layout(legend_title_text="Score components")
newnames = {'bad_afin':'Invalid Avidity', 
            'correctness': 'Correctness', 
            "coverage": "Coverage", 
            "f1": "f1", 
            "good_afin": "Valid Avididy",
            "precession": "Precision",
            "recall": "Recall",
            "uniqueness": "Uniqueness",
            }
bad_names = ["f1", "Precision", "Recall"]
fig_agg_vals.for_each_trace(lambda t: t.update(name = newnames[t.name],
                                      legendgroup = newnames[t.name]
                                    #   hovertemplate = t.hovertemplate.replace(t.name, newnames[t.name])
                                     )
                  )

fig_agg_vals.for_each_trace(lambda t: t.update(visible=False) if t.name in bad_names else ())

write_image(fig_agg_vals, f"{file_pre}", f"agg_ScoreComponents.png")
fig_agg_vals.show()
fig_pop_dim_vals = plot_agg_vals(df_list, "PopDimTypeMemberships_", file_pre, sample_rate=1)
fig_pop_dim_vals.update_layout(legend = dict(font = dict(size = 18, color = "black")),
                           legend_title = dict(font = dict(size = 18)),
                           legend_title_text = "Dimension type")
write_image(fig_pop_dim_vals, f"{file_pre}", f"agg_PopDimTypeMemberships.png")
fig_pop_dim_vals.show()
fig_pop_label = plot_agg_vals(df_list, "PopLabelMemberships_", file_pre, first_n=100)
write_image(fig_pop_label, f"{file_pre}", f"agg_PopLabelMemberships.png")
fig_pop_label.show()

avg_train_score = plot_agg_vals(df_list, "AvgTrainScore", file_pre, first_n=100)
write_image(avg_train_score, f"{file_pre}", f"agg_AvgTrainScore.png")
avg_train_score.show()

boosting_acc = plot_agg_vals(df_list, "BoostAccuracy", file_pre, first_n=300)
write_image(boosting_acc, f"{file_pre}", f"agg_BoostAccScore.png")
boosting_acc.show()

boosting_acc = plot_agg_vals(df_list, "BoostAccuracyTest", file_pre, first_n=300)
write_image(boosting_acc, f"{file_pre}", f"agg_BoostAccScore.png")
boosting_acc.show()

avg_train_acc = plot_agg_vals(df_list, "Accuracy", file_pre, first_n=100)
newnames = {'racy_accuracy':'Test accuracy', 
            'uracy_accuracy': 'Train accuracy',
            "racy_cor":"racy_cor",
            "racy_no_reg":"racy_no_reg",
            "racy_wrong":"racy_wrong",
            "uracy_cor":"uracy_cor",
            "uracy_no_reg":"uracy_no_reg",
            "uracy_wrong":"uracy_wrong",
            "racy_precision":"racy_precision",
            "uracy_precision":"uracy_precision"
            }
bad_names = ["racy_cor", "racy_no_reg", "racy_wrong", "racy_precision", "uracy_cor", "uracy_no_reg", "uracy_wrong", "uracy_precision"]
avg_train_acc.for_each_trace(lambda t: print(t.name))
avg_train_acc.for_each_trace(lambda t: t.update(name = newnames[t.name],
                                       legendgroup = newnames[t.name]
                                     #   hovertemplate = t.hovertemplate.replace(t.name, newnames[t.name])
                                      )
                   )
avg_train_acc.for_each_trace(lambda t: t.update(visible=False) if t.name in bad_names else ())
avg_train_acc.update_yaxes(range=[0.6, 1.001])
write_image(avg_train_acc, f"{file_pre}", f"agg_TrainAccuracy.png")
avg_train_acc.show()

# avg_test_acc = plot_agg_vals(df_list, "TestAccuracy", file_pre, first_n=100)
# write_image(avg_test_acc, f"{file_pre}", f"agg_TestAccuracy.png")
# avg_test_acc.show()

for n, df in enumerate(df_list):
    continue
    print(f"Fold {n}:")
    plot_train_test_acc(df, n)
    plot_score_comp_memberships(df.iloc[::sample_rate, :], n, "ScoreComponents_")
    plot_score_comp_memberships(df.iloc[::sample_rate, :], n, "PopLabelMemberships_")
    plot_score_comp_memberships(df.iloc[::sample_rate, :], n, "PopDimTypeMemberships_")

    fig = px.line(df["AvgTrainScore"].ffill())
    fig.update_layout(height = 500)
    fig.show()
```
