```python
#with open("./../../ais/logg_dat/logg_dat_Diabetes.json", "r") as f:
with open("./../../ais/logg_dat/logg_dat_EmbeddingBuzfeed.json", "r") as f:

#with open("./../../ais/logg_dat/logg_dat_Iris.json", "r") as f:
    js_obj = json.load(f)

df_list = parse_iter_props(js_obj)

for df in df_list:
    gen_cor_w_noreg_cols(df, "TestAccuracy")
    gen_cor_w_noreg_cols(df, "TrainAccuracy")

def plot_train_test_acc(ddf: pd.DataFrame):
    fig = go.Figure()
    fig.add_trace(go.Scatter(x=ddf.index, y=ddf["TestAccuracy_accuracy"].ffill(),
                        mode='lines',
                        name='TestAccuracy_accuracy'))
    
    fig.add_trace(go.Scatter(x=ddf.index, y=ddf["TrainAccuracy_accuracy"].ffill(),
                        mode='lines',
                        name='TrainAccuracy_accuracy'))
    
    fig.update_layout(height = 500)
    fig.show()
    
def plot_score_comp_memberships(ddf: pd.DataFrame, prefix: str):
    cols = ddf.columns
    fig = go.Figure()

    for col in cols:
        if col.startswith(prefix):
            fig.add_trace(go.Scatter(x=ddf.index, y=ddf[col].ffill(),
                                mode='lines',
                                name=col))

    fig.update_layout(height = 500)
    fig.show()


for n, df in enumerate(df_list):
    print(f"Fold {n}:")
    plot_train_test_acc(df)
    plot_score_comp_memberships(df, "ScoreComponents_")
    plot_score_comp_memberships(df, "PopLabelMemberships_")
    plot_score_comp_memberships(df, "PopDimTypeMemberships_")

    fig = px.line(df["AvgTrainScore"].ffill())
    fig.update_layout(height = 500)
    fig.show()
```
